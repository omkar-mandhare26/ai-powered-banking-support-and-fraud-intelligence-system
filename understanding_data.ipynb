```python
st_cols = support_tickets.columns

for col in st_cols:
    print("--------------------------------------------")
    print(f"Column: {col}")
    print(support_tickets[col].unique())
    print()
```

```python
support_tickets["amt"] = (
    support_tickets["query_text"]
    .str.extract(r"₹([\d,]+)")[0]
    .str.replace(",", "", regex=False)
)

support_tickets["sub_category"] = support_tickets.apply(
    lambda row: row["sub_category"].replace("{amt}", row["amt"])
    if pd.notna(row["amt"]) else row["sub_category"],
    axis=1
)

support_tickets = support_tickets.drop(columns=["amt"])

support_tickets
```
